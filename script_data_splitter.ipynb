```python
import pandas as pd
import os
```

```python
file_path = input('Veulliez entrer le nom du fichier: ')
file_path += '.csv'
if os.path.isfile(file_path):
    with open(file_path, 'r') as f:
        df = pd.read_csv(file_path)
else:
    print(f"The file '{file_path}' does not exist.")
```

```python
df_clean = df.dropna()
df_clean = df_clean.sample(frac=1).reset_index(drop=True)
df_clean
```

```python
json_perc = 0
db_perc = 0

while True:
    json_perc = int(input('Veuillez saisir le pourcentage du fichier json: '))
    if json_perc < 1 or json_perc > 100:
        print("EntrÃ©e non valide ! Veuillez saisir un pourcentage entre 1 et 100.")
    else:
        break

while True:
    db_perc = int(input('Veuillez saisir le pourcentage de la base de donnees: '))
    if db_perc < 1 or db_perc >= 100-json_perc:
        print(f"EntrÃ©e non valide ! Veuillez saisir un pourcentage entre 1 et {100-json_perc}.")
    else:
        break

njs = len(df_clean) * (json_perc / 100)
ndb = len(df_clean) * (db_perc / 100)
ncs = len(df_clean) * ((json_perc + db_perc) / 100)

df_js = df_clean[:round(njs)]
df_db = df_clean[round(ndb):round(ncs)]
df_csv = df_clean[round(ncs):]

print(f"Le nombre de ligne du fichier json est {len(df_js)} pour le pourcentage {json_perc}% ")
print(f"Le nombre de ligne de la base de donnees est {len(df_db)} pour le pourcentage {db_perc}% ")
print(f"Le nombre de ligne du fichier csv est {len(df_csv)} pour le pourcentage {(100 - (json_perc + db_perc))}% ")
```

```python
# Prompt the user to enter the name of a JSON file
while True:
    nom_json = input('Veuillez saisir le nom du fichier json: ')
    if not nom_json.isidentifier():
        print('Le nom du fichier json doit Ãªtre un identifiant valide.')
    else:
        break

# Prompt the user to enter the name of a database
while True:
    nom_db = input('Veuillez saisir le nom de la base de donnÃ©es: ')
    if not nom_db.isidentifier():
        print('Le nom de la base de donnÃ©es doit Ãªtre un identifiant valide.')
    else:
        break

# Prompt the user to enter the name of a table in the database
while True:
    nom_table = input('Veuillez saisir le nom de la table de votre base de donnÃ©es: ')
    if not nom_table.isidentifier():
        print('Le nom de la table doit Ãªtre un identifiant valide.')
    else:
        break

print('Les noms de fichier JSON, de base de donnÃ©es et de table sont valides!')

df_js.to_json(f'{nom_json}.json', orient='records')
df_csv.to_csv(f'{nom_json}-split.csv', header=False, index=False)
print('le fichier JSON et le fichier CSV sont bien cree!')

```

```python
import pyodbc as odbc

# Set up the connection string
driver = 'SQL SERVER'
server = 'PC-HICHAM'
database = 'master'

connection_info = f"""
    DRIVER={{{driver}}};
    SERVER={{{server}}};
    DATABASE={{{database}}};
    Trust_Connection=yes;
"""
# Connect to default database
connection = odbc.connect(connection_info)
cursor = connection.cursor()
```

```python
# create database and use it
cursor.execute(f'''
    IF EXISTS (SELECT name FROM sys.databases WHERE name = '{nom_db}')
        PRINT 'Database {nom_db} exist'
    ELSE
        CREATE DATABASE {nom_db};
''')
#commit changes
connection.commit()
```

```python
cursor.execute(f''' USE {nom_db} ''')
#commit changes
connection.commit()
```

```python
# Function to get the SQL Server data type equivalent for DataFrame types
def get_sql_server_type(ctype):
    sql_server_types = {
        'int64': 'INT',
        'object': 'NVARCHAR(max)',
        'bool': 'BIT',
        'float64': 'FLOAT'
    }
    return sql_server_types.get(ctype, 'NVARCHAR(max)')

# Generate the SQL Server column definitions based on DataFrame types
sql_server_columns = ',\n'.join([f'{column} {get_sql_server_type(str(df_db.dtypes[column]))}' for column in df_db.columns])

# SQL Server CREATE TABLE statement
create_table_sql = f''' CREATE TABLE {nom_table} ( {sql_server_columns});'''

# print(create_table_sql)
cursor.execute(create_table_sql)
connection.commit()
```

```python
column_names = ', '.join(df_db.columns)

value_placeholders = ', '.join(['?' for _ in df_db.columns])

# Generate the SQL Server INSERT INTO statement
insert_query = f"INSERT INTO {nom_table} ({column_names}) VALUES ({value_placeholders})"
cursor.cancel()
# Execute the INSERT INTO statement for each row of data in the DataFrame
for _, row in df_db.iterrows():
    values = tuple(row.values)
    cursor.execute(insert_query, values)
    print(f'success insertion of row {values} .')

# Commit the changes and close the connection
connection.commit()
```
